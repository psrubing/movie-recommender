```python
import numpy as np
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv('ratings.csv', sep=',')
ratings.drop('timestamp',axis=1,inplace=True)
ratings.head()
```

```python
movies = pd.read_csv('movies.csv',sep=',')
movies.drop(columns='genres',inplace=True)
movies.head()
```

```python
df = pd.merge(ratings,movies,on='movieId')
df.head()
```

```python
ratings_overall = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings_overall['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings_overall.head()
```

```python
sns.distplot(ratings_overall['num of ratings'],bins=200,kde=False)
plt.xlim(0,60)
```

```python
pivot_movie = df.pivot_table(index='title',columns='userId', values='rating')
pivot_movie = pivot_movie.dropna(thresh=5,axis=1).fillna(0)
pivot_movie
```

```python
cos_mat = cosine_similarity(pivot_movie)
cos_mat.shape
```

```python
cos_mat_df = pd.DataFrame(cos_mat,columns=pivot_movie.index,index=pivot_movie.index)
cos_mat_df.head()
```
